```python
import numpy as np
import matplotlib.pyplot as plt

def f(r,t):
    x, y = r
    fx = x*y - x
    fy = y - x*y + np.sin(t)**2
    return np.array([fx,fy], float)

a, b = 0.0, 10.0
N = 100
h = (b-a)/N

tpoints = np.arange(a,b,h)
r = np.zeros((N,2))
r[0] = [1.0, 1.0]

# Loop over time to solve for x and y, which will be stored in r[]
for i,t in enumerate(tpoints[1:], 1):
    r_i = r[i-1]
    k1 = h * f(r_i, t)
    k2 = h * f(r_i + 0.5*k1, t + 0.5*h)
    k3 = h * f(r_i + 0.5*k2, t + 0.5*h)
    k4 = h * f(r_i + k3, t + h)
    r[i] = r_i + (k1 + 2*k2 + 2*k3 + k4) / 6.0

plt.plot(r[:,0], r[:,1])
plt.xlabel("t")
plt.ylabel("x,y")
plt.show()
```
