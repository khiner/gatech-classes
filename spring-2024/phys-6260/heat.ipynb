```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
L = 0.01      # Thickness of steel in meters
D = 4.25e-6   # Thermal diffusivity
N = 100       # Number of divisions in grid
a = L/N       # Grid spacing
h = 1e-4      # Time-step
epsilon = h/1000

Tlo = 0.0     # Low temperature in Celcius
Tmid = 20.0   # Intermediate temperature in Celcius
Thi = 80.0    # Hi temperature in Celcius

t1 = 0.01
t2 = 0.1
t3 = 0.4
t4 = 1.0
t5 = 10.0
tend = t5 + epsilon

# Create arrays
T = np.zeros(N+1,float)
T[0] = Thi
T[N] = Tlo
T[1:N] = Tmid
Tp = np.zeros(N+1,float)
Tp[0] = Thi
Tp[N] = Tlo

# Main loop
t = 0.0
c = h*D/(a*a)
all_times = [t1, t2, t3, t4, t5]
all_is = [int(t/h) for t in all_times]
i = 0
while t<tend:
    # In-class problem
    #
    # Calculate the new values of T
    Tp[1:N] = T[1:N] + c * (T[2:N+1] + T[:N-1] - 2 * T[1:N])
    T, Tp = Tp, T
    # In-class problem
    #
    # Make plots at the given times (t1 .. t5)
    if i in all_is:
        plt.plot(Tp)
    i += 1
    t += h

plt.xlabel("x/L")
plt.ylabel("T")
plt.legend(loc="best")
plt.savefig("20_heat.png")
```
