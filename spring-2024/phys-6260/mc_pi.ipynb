```python
import numpy as np
import matplotlib.pyplot as plt


# Estimate pi/4 using Monte Carlo

max_N = int(1e4)
estimates = []
for N in range(1, max_N):
    # Generate N random numbers in the interval [[0, 0], [1, 1]].
    test_values = np.random.rand(N, 2)
    # Test if the point is inside the unit circle.
    num_inside = np.sum(np.sum(test_values**2, axis=1) < 1)
    estimates.append(num_inside / N)

plt.title('Estimate of $\\pi/4$ using Monte Carlo')
plt.xlabel('N')
plt.ylabel('Estimate of $\\pi/4$')
plt.plot(estimates, label='Estimate', lw=0.5)
plt.hlines(np.pi/4, 0, max_N, color="r", label="True value", lw=2)
plt.xscale('log')
plt.autoscale(tight=True)
plt.grid()
plt.legend()
```
